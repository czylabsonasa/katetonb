## Create networks

```r
rm(list = ls()) # Remove all the objects we created so far.
library(igraph) # Load the igraph package
# adjust the size
library(repr)
options(repr.plot.width=6, repr.plot.height=4) # adjust these numbers according your taste
```

The code below generates an undirected graph with three edges. 
The numbers are interpreted as vertex IDs, so the edges are `1–>2, 2–>3, 3–>1`.

```r
g1 <- graph( edges=c(1,2, 2,3, 3, 1), n=3, directed=F ) 
class(g1)
g1
plot(g1) # A simple plot of the network - we'll talk more about plots later
```

```r
# Now with 10 vertices, and directed by default:
g2 <- graph( edges=c(1,2, 2,3, 3, 1), n=10 )
g2
class(g2)
plot.igraph(g2)   
```

```r
g3 <- graph( c("John", "Jim", "Jim", "Jill", "Jill", "John")) # named vertices
# When the edge list has vertex names, the number of nodes is not needed
g3
plot(g3)
```

```r
g4 <- graph( c("John", "Jim", "Jim", "Jack", "Jim", "Jack", "John", "John"), 
             isolates=c("Jesse", "Janis", "Jennifer", "Justin") )  
# In named graphs we can specify isolates by providing a list of their names.
plot(g4, edge.arrow.size=.5, vertex.color="gold", vertex.size=10, 
     vertex.frame.color="gray", vertex.label.color="black", 
     vertex.label.cex=0.8, vertex.label.dist=4, edge.curved=0.5) 
```

Small graphs can also be generated with a description of this kind: `-` for undirected tie, `+-` or `-+` for directed ties pointing left or right, `++` for a symmetric tie, and `:` for sets of vertices.

```r
plot(graph_from_literal(a---b, b---c)) # the number of dashes doesn't matter
```

```r
plot(graph_from_literal(a--+b, b+--c))    
```

```r
plot(graph_from_literal(a+-+b, b+-+c)) 
```

```r
gl <- graph_from_literal(a-b-c-d-e-f, a-g-h-b, h-e:f:i, j)
plot(gl)
```
