## Specific graphs and graph models

```r
library(igraph)
# adjust these numbers according your taste,
# the default plotting sizes are too big in my browser
options(repr.plot.width=10, repr.plot.height=5) 
```

### Empty graph

```r
eg <- make_empty_graph(40)
plot(eg, vertex.size=10, vertex.label=NA)
```

### Full (complete) graph

```r
fg <- make_full_graph(8)
plot(fg, vertex.size=10, vertex.label=NA)
```

### Simple star graph

```r
st <- make_star(11)
plot(st, vertex.size=10, vertex.label=NA) 
```

### Tree graph

```r
tr <- make_tree(40, children = 3, mode = "undirected")
plot(tr, vertex.size=10, vertex.label=NA) 
```

### Ring graph

```r
rn <- make_ring(40)
plot(rn, vertex.size=10, vertex.label=NA)
```

### Erdős-Rényi random graph model

```r
# play with various values of m
er <- sample_gnm(n=100, m=200) 
plot(er, vertex.size=6, vertex.label=NA,edge.color="black") 
```

### Watts-Strogatz small-world model

```r
# for interpreting the parameters: see the help (?sample_smallworld)
sw <- sample_smallworld(dim=2, size=5, nei=1, p=0.1)
plot(sw, vertex.size=6, vertex.label=NA, layout=layout_in_circle)
```

### Barabási-Albert preferential attachment model for scale-free graphs<br>
(n is number of nodes, power is the power of attachment (1 is linear); m is the number of edges added on each time step)

```r
ba <-  sample_pa(n=50, power=1, m=2,  directed=F)
plot(ba, vertex.size=6, vertex.label=NA,edge.color = "black")
```

`igraph` can also give you some notable historical graphs. For instance:

```r
zach <- graph("Zachary") # the Zachary carate club
plot(zach, vertex.size=6, vertex.label=NA,edge.color="black")
```

### Rewiring a graph<br>
`each_edge()` is a rewiring method that changes the edge endpoints uniformly randomly with a probability `prob`.


```r
# rewire the previously defined ring graph rn
rn.rewired <- rewire(rn, each_edge(prob=0.3))
plot(rn.rewired, vertex.size=7, vertex.label=NA,layout=layout.circle)
diameter(rn.rewired)
diameter(rn)
```

```r
# combine the previous tree and ring 
plot(rn %du% tr, vertex.size=7, vertex.label=NA) 
```
